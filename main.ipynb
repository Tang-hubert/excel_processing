```python

for i in range(uni_information.shape[0]):
    for j in range(OUT_ADMISSION_103_106.shape[0]):
        if (uni_information['School_NAME'][i] == OUT_ADMISSION_103_106['uni_school_name'][j] and uni_information['DEPT_NAME'][i] == OUT_ADMISSION_103_106['department_name'][j]):
            OUT_ADMISSION_103_106['DEPT_TYPE'][j] = uni_information['DEPT_TYPE'][i]
            OUT_ADMISSION_103_106['AREA'][j] = uni_information['AREA'][i]
        else: j+=1
```
